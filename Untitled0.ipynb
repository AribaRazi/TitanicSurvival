```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df=sns.load_dataset("titanic") # already built in dataset examples in sns
```

```python
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
df = df.drop(columns=['alive','embark_town','deck','adult_male','who','class'])
```

```python
# age has missing values so filling those null values
```

```python
df["age"].fillna(df["age"].mean(),inplace=True) #891--> 889 for all columns
```

```python
df.info()
```

```python
df.dropna(subset=["embarked"],inplace=True) #891--> 889 for all columns
```

```python
df.head()
```

# label encoding


```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df.head()
```

```python
df['sex']=le.fit_transform(df['sex'])
df['embarked']=le.fit_transform(df['embarked'])
```

```python
df.head()
df=df.astype(int)
```

```python
df.head()
```

## splitting into X and y variables

```python
X=df.drop("survived",axis=1)
y=df["survived"]
```

```python
# train test and split
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
# now we willl compare y_pred with y_test(original value)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
